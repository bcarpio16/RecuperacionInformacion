#### Matriz TF-IDF

```python
import os
import unicodedata
import string
import math
from tabulate import tabulate

dataset_path = '/kaggle/input/dataset1'

terminos_unicos = set()
documentos = []           

for archivo in os.listdir(dataset_path):
    archivo_path = os.path.join(dataset_path, archivo)

    with open(archivo_path, 'r', encoding='utf-8') as file:
        lineas = file.readlines()
        for linea in lineas:
            palabras_doc = []
            linea = linea.strip().lower()
            palabras = linea.split()

            for palabra in palabras:
                palabra = unicodedata.normalize('NFD', palabra)
                palabra = ''.join(c for c in palabra if unicodedata.category(c) != 'Mn')
                palabra = palabra.translate(str.maketrans('', '', string.punctuation))

                if palabra:
                    terminos_unicos.add(palabra)
                    palabras_doc.append(palabra)

            if palabras_doc:
                documentos.append(palabras_doc)

print(f"Número de palabras únicas: {len(terminos_unicos)}")
print(f"Número de documentos: {len(documentos)}")
```

```python
documentos_texto = [' '.join(doc) for doc in documentos]

def crear_matriz_terminos(texto_docs):
    todos_terminos = set()
    for doc in texto_docs:
        todos_terminos.update(doc.split())

    lista_terminos = sorted(todos_terminos)
    matriz = []

    for termino in lista_terminos:
        fila = []
        for doc in texto_docs:
            fila.append(1 if termino in doc.split() else 0)
        matriz.append(fila)

    return lista_terminos, matriz

def calcular_idf(terminos, matriz):
    num_docs = len(matriz[0])
    idf_valores = {}
    for i, termino in enumerate(terminos):
        df = sum(matriz[i])
        idf_valores[termino] = round(math.log10(num_docs / (df)),3)
    return idf_valores

def mostrar_matriz(terminos, matriz, idf_valores):
    tabla = []
    for i, termino in enumerate(terminos):
        fila = [termino] + matriz[i] + [idf_valores[termino]]
        tabla.append(fila)

    headers = ["Término"] + [f"Doc{i+1}" for i in range(len(matriz[0]))] + ["IDF"]
    print(tabulate(tabla, headers=headers, tablefmt="grid", showindex=False))

terminos, matriz = crear_matriz_terminos(documentos_texto)
idf_valores = calcular_idf(terminos, matriz)
mostrar_matriz(terminos, matriz, idf_valores)
```
